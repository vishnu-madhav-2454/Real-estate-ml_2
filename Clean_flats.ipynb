```python
import numpy as np
```

```python
import pandas as pd
```

```python
df = pd.read_csv('flats.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(['link','property_id'],axis = 1)
```

```python
df.rename(columns={'area':'price_per_sqft'},inplace=True)
```

```python
df['society'].value_counts()
```

```python
df['society'] = df['society'].str.lower()
```

```python
df['society'].value_counts()
```

```python
df = df[df['price'] != "Price on Request"]
```

```python
def Crore_into_lac(price):
  if isinstance(price,str):
    if 'Crore' in price:
      return float(price.split(' ')[0])*100
    else:
      return float(price.split(' ')[0])
df['price'] = df['price'].apply(Crore_into_lac)
```

```python
def remove_sq_ft(x):
  if isinstance(x,str):
    val = x.split(' ')[1].split('/')[0].split(',')
    if len(val) == 1:
      return float(val[0])
    else:
      return float(val[0])*1000+float(val[1])
df['price_per_sqft'] = df['price_per_sqft'].apply(remove_sq_ft)
```

```python
df['bedRoom'] = df['bedRoom'].str.replace('Bedrooms','').str.replace('Bedroom','').astype(float)
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].str.replace('Bathrooms','').str.replace('Bathroom','').astype(float)
```

```python
df['balcony'].value_counts()
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df['balcony'] = df['balcony'].str.replace('Balconies','').str.replace('Balcony','')
```

```python
df['additionalRoom'].fillna('notavailable',inplace=True)
```

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['additionalRoom'].value_counts()
```

```python
df['floorNum'].value_counts()
```

```python
def find_floor_num(x):
  if isinstance(x,str):
    num = x.split(" ")
    if num[0] == 'Ground':
      return 0
    elif num[0] == 'Basement':
      return -1
    elif num[0] == 'Lower':
      return 0
    elif 'nd' in num[0]:
      return float(num[0].split('n')[0])
    elif 'rd' in num[0]:
      return float(num[0].split('r')[0])
    elif 'th' in num[0]:
      return float(num[0].split('t')[0])
    elif 'st' in num[0]:
      return float(num[0].split('s')[0])
    else:
      return float(num[0])
df['floorNum'] = df['floorNum'].apply(find_floor_num)
```

```python
df['facing'].fillna('NA',inplace=True)
```

```python
df['area'] = (df['price'] * 100000)/df['price_per_sqft']
```

```python
df.head()
```

```python
df.to_csv('flats_cleaned.csv',index=False)
```
